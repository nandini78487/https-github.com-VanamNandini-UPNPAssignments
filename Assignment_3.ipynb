1.Generate 10000 values between 5 and 500.Find its mean,meadian, variance,25th percentile.

```python
a=np.random.randint(5,500,10000)
print(a)
print(np.mean(a))
print(np.median(a))
print(np.var(a))
print(np.percentile(a,25))
```
